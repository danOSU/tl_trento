```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
D2_trento_design = pd.read_csv('2D_500POINTS_rescaled.txt', header=0, names=['N', 'w', r'$\chi$'])
D3_trento_design = pd.read_csv('3D_150POINTS_rescaled.txt', header=0, names=['N', 'w', r'$\chi$', r'k$_{T,min}$', 
                                                                             r'$\alpha$', r'$\beta$', 'flatness'])
D3_trento_test_design = pd.read_csv('Test_3D_300POINTS_rescaled.txt', header=0, names=['N', 'w', r'$\chi$', r'k$_{T,min}$', 
                                                                             r'$\alpha$', r'$\beta$', 'flatness'])
```

```python
# Check for nested design
diff = D2_trento_design.iloc[0:150,:] - D3_trento_design.iloc[0:150,0:3]
if np.all(np.isclose(diff,0)):
    print('All good, The first 150 design points of 2d Trento and 3D trento are nested')
    print('Diference matrix:')
    print(diff)
```
